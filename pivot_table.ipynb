```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv('world-happiness-report.csv')
df.head()
```

```python
df_pivot = df.pivot_table(index = 'Country name', columns = 'year', values = 'Freedom to make life choices').fillna(0)
df_pivot
```

```python
from scipy.sparse import csr_matrix
df_pivot_matrix = csr_matrix(df_pivot.values)
print(df_pivot_matrix)
```

Fitting Data into NearestNeighborsModel

```python
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric = 'cosine', n_neighbors = 20, radius = 1)
model_knn.fit(df_pivot_matrix)
```

Matrix of Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(df_pivot)
similarity_matrix
```

```python
Country_name = (input("Enter Country Name according to the Data Set: "))
data = list(df_pivot.index)
print(data)
```

```python
query_index = data.index(Country_name)
print(query_index)
```

```python
similarity, indices = model_knn.kneighbors(df_pivot.iloc[query_index,:].values.reshape(1, -1), 
                                           n_neighbors = 8)
print(similarity)
print(indices)
```

Create Dictionary and store recommendations of Country_name
Which is given by USER

```python
data_dict = {}
for i in range(0, len(similarity.flatten())):
    if i == 0:
        print('Recommendation for {0}: \n'.format(df_pivot.index[query_index]))
    else:
        data_dict[str(df_pivot.index[indices.flatten()[i]])] = float(similarity.flatten()[i])
        print(f'{df_pivot.index[indices.flatten()[i]]}, has similarity with {similarity.flatten()[i]: .20f}: ')
print(data_dict)
```
